```python
from itertools import permutations  
def permutation_sample(num_fc_cases,num_mc_cases):
    """Generate a permutation sample from the 2 data sets"""
    c_both_LV = np.concatenate((num_fc_cases,num_mc_cases), axis=None)
    c_both_LV_perm= np.random.permutation(c_both_LV)
    perm_sample_FLV = c_both_LV_perm[:len(num_fc_cases)]
    perm_sample_MLV = c_both_LV_perm[len(num_fc_cases):]
    
```

Since we are getting a strange histogram, the sample size was >30 and thus the next step will be running a parametric test since we cannot increase or sample size. Thus onward to a 2 sample t test.


```python
#Null hypothesis significance testing
def draw_perm_reps(num_fc_cases,num_mc_cases, func, size=1):
    """Generate multiple permutation replicates"""
    perm_replicates=np.empty(size)
    
    for i in range(size):
        perm_sample_1, perm_sample_2 = permutation_sample(num_fc_cases,num_mc_cases)
        
        perm_replicates[i]= func(perm_sample_1, perm_sample_2)
        
    return perm_replicates    
    
   
```

```python
def diff_of_means(num_fc_cases,num_mc_cases):
    """Difference of means of 2 arrays"""
    diff=np.mean(num_fc_cases)-np.mean(num_mc_cases) 
    return diff
```

```python
empirical_diff_means = diff_of_means(num_fc_cases,num_mc_cases)
perm_replicates = draw_perm_reps(num_fc_cases,num_mc_cases, diff_of_means, size=10000)
p=np.sum(perm_replicates >= empirical_diff_means)/ len(perm_replicates)
print('p-value is: ', p)
```
